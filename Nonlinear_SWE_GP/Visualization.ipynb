```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
true_predictions = np.load('PCA_Coefficients_q1_test.npy')
pred_predictions = np.load('PCA_Coefficients_q1_pred.npy')
```

```python
mode_num = 0
for i in range(10):
    plt.figure()
    plt.plot(true_predictions[mode_num,i*10:(i+1)*10],label='True')
    plt.plot(pred_predictions[mode_num,i*10:(i+1)*10],label='Predicted')
    plt.legend()
    plt.title('Test simulation '+str(i))
    plt.show()
```
