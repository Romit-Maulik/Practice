```python
import numpy as np
np.random.seed(10)
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def func(y,t, amat, b, c, k, n_units):
    x_= y[0:n_units]
    y_= y[n_units:]
    
    dydt = []
    
    for i in range(n_units):
        temp_1 = x_[i]*(a[i] - x_[i])*(x_[i] - 1.0) - y_[i]
        for j in range(n_units):
            temp_1 = temp_1 + k*amat[i,j]*(x_[j]-x_[i])
            
        dydt.append(temp_1)
        
    for i in range(n_units):
        temp_2 = b[i]*x_[i] - c[i]*y_[i]
        
        dydt.append(temp_2)
        
    return dydt
```

```python
n_units = 2
a = np.zeros(shape=(n_units))
b = np.zeros(shape=(n_units))
c = np.zeros(shape=(n_units))
amat = np.zeros(shape=(n_units,n_units))


a[:] = -0.025794
b[0] = 0.0065
b[1] = 0.0135
c[:] = 0.02
k = 0.128

amat[0,1] = 1
amat[1,0] = 1

params = [amat, b, c, k, n_units]
```

```python
y0 = [0.2,0.2,0.0,0.0]
t = np.linspace(0,50000,500000)
```

```python
sol = odeint(func, y0, t, args=(amat, b, c, k, n_units))
```

```python
plt.plot(t, sol[:, 0], 'b', label='x1')
plt.plot(t, sol[:, 1], 'g', label='x2')
plt.plot(t, sol[:, 2], 'r', label='y1')
plt.plot(t, sol[:, 3], 'y', label='y2')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
plt.plot(t, (sol[:, 0]+sol[:, 1])/2.0, 'b', label='x_average')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
