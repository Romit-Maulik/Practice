```python
import numpy as np
from arviz import *
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(10)
from pymc3 import *
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read data file 
choi_df = pd.read_csv('Recoded_data_v2.csv',encoding = "ISO-8859-1")
choi_df = choi_df.fillna(0.0)
data = np.asarray(choi_df.values.tolist())

choi_df.head()
```

```python
# Move data to numpy arrays
data = np.asarray(choi_df.values.tolist())
# Ready to do operations
independent_vars = data[:,:-1].astype('double') # Last column is dependent
dependent_vars = data[:,-1].astype('double')

# # Do normalization of features
min_max_ip_scaler = MinMaxScaler()
independent_vars = min_max_ip_scaler.fit_transform(independent_vars)
```

```python
with Model() as model: # model specifications in PyMC3 are wrapped in a with-statement
    # Define priors
    sigma = HalfCauchy('sigma', beta=10, testval=1.)
    intercept = Normal('Intercept', 0, sigma=20)
    
    coeff_1 = Normal('c1', 0, sigma=20)
    coeff_2 = Normal('c2', 0, sigma=20)
    coeff_3 = Normal('c3', 0, sigma=20)
    coeff_4 = Normal('c4', 0, sigma=20)
    coeff_5 = Normal('c5', 0, sigma=20)
    coeff_6 = Normal('c6', 0, sigma=20)
    coeff_7 = Normal('c7', 0, sigma=20)
    coeff_8 = Normal('c8', 0, sigma=20)
    coeff_9 = Normal('c9', 0, sigma=20)
    coeff_10 = Normal('c10', 0, sigma=20)
    coeff_11 = Normal('c11', 0, sigma=20)
    coeff_12 = Normal('c12', 0, sigma=20)
    coeff_13 = Normal('c13', 0, sigma=20)
    coeff_14 = Normal('c14', 0, sigma=20)
    coeff_15 = Normal('c15', 0, sigma=20)
    coeff_16 = Normal('c16', 0, sigma=20)

    # Define likelihood
    likelihood = Normal('op', mu=intercept + coeff_1*independent_vars[:,0] + coeff_2*independent_vars[:,1] + coeff_3*independent_vars[:,2]
                        + coeff_4*independent_vars[:,3] + coeff_5*independent_vars[:,4] + coeff_6*independent_vars[:,5] + coeff_7*independent_vars[:,6]
                         + coeff_8*independent_vars[:,7] + coeff_9*independent_vars[:,8] + coeff_10*independent_vars[:,9] + coeff_11*independent_vars[:,10]
                         + coeff_12*independent_vars[:,11] + coeff_13*independent_vars[:,12] + coeff_14*independent_vars[:,13] + coeff_15*independent_vars[:,14]
                         + coeff_16*independent_vars[:,15],
                        sigma=sigma, observed=dependent_vars)

    # Inference!
    trace = sample(3000, cores=2) # draw 3000 posterior samples using NUTS sampling
    
```

```python
plt.figure(figsize=(7, 7))
traceplot(trace[100:])
plt.tight_layout();
```
