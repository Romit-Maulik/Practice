```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
ds_train = xr.open_dataset('http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.oisst.v2/sst.wkmean.1981-1989.nc')
ds_test = xr.open_dataset('http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.oisst.v2/sst.wkmean.1990-present.nc')
ds_mask = xr.open_dataset('http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.oisst.v2/lsmask.nc')
```

```python
# Inspect train data
ds_train
```

```python
print(ds_train.sst.attrs)
```

```python
lat_grid = np.asarray(ds_train.lat)
lon_grid = np.asarray(ds_train.lon)
```

```python
# Example plot using xr
ds_train.sst[0].plot()
```

```python
ds_mask.mask.attrs
```

```python
# Plot with mask
(ds_train.sst[0]*ds_mask.mask[0]).plot()
```

```python
# Extract snapshots to numpy arrays - this actually downloads
ds_train_np = ds_train.sst[:]*ds_mask.mask[0]
ds_test_np = ds_test.sst[:]*ds_mask.mask[0]

print(np.shape(ds_train_np))
```

```python
ds_train_np = np.asarray(ds_train_np)
ds_test_np = np.asarray(ds_test_np)
```

```python
# Test plotting of numpy arrays
fig, ax = plt.subplots(nrows=1,ncols=1)
levels = np.linspace(-5.0, 35.0, num=9)
cx0 = ax.contourf(lon_grid,lat_grid,ds_train_np[100],levels=levels)
ax.set_title('NOAA SST (numpy)')
ax.set_ylabel('Latitude')
ax.set_xlabel('Longitude')
fig.colorbar(cx0,ax=ax)
plt.show()
```

```python
# Save to disk so you don't have to reload everytime (internet downloads are expensive)
np.save('NOAA_Train_Snapshots.npy',ds_train_np)
np.save('NOAA_Test_Snapshots.npy',ds_test_np)
```

```python
# Save the mask
mask = np.asarray(ds_mask.mask)
np.save('NOAA_Mask.npy',mask)
```
